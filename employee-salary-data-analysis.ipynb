```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
from tkinter import *
df=pd.read_csv('C:/Users/samee/OneDrive/Desktop/Employee_Salaries.csv')
df['Net_Pay']=df['Base_Salary']+df['Overtime_Pay']+df['Longevity_Pay']
df.head()
import tkinter as tk

# Create the main window
root = tk.Tk()
root.title("Main Page")
root.geometry("600x350")

# Create a label widget
label = tk.Label(root, text="Employee Salaries DataSet Analysis",font=('times',18,'bold'))

# Displaying the label on the screen
label.pack()

# Creating the 4 buttons
button1 = tk.Button(root, text="Bar Graph to show the Net Pay of Men & Women", command=graph1,font=('times',14,'bold'))
button2 = tk.Button(root, text="Distribution of Net Pay in the form of a Density Plot",command=graph2,font=('times',14,'bold'))
button3 = tk.Button(root, text="Scatter Plot of Base Salary and Net Pay of the Employees",command=graph3,font=('times',14,'bold'))
button4 = tk.Button(root, text="Histogram of Base Salary of Employees",command=graph4,font=('times',14,'bold'))

# Pack the buttons into the window
button1.pack(pady=20)
button2.pack(pady=20)
button3.pack(pady=20)
button4.pack(pady=20)

# Start the main event loop to display the window
root.mainloop()

def graph1():
    sns.barplot(df,x='Gender',y='Net_Pay')
    plt.show()
def graph2():
    sns.kdeplot(df['Net_Pay'],fill=True)
    plt.show()
def graph3():
    sns.regplot(df,x='Base_Salary',y='Net_Pay')
    plt.show()
def graph4():
    sns.histplot(df['Base_Salary'])
    plt.show()
```
